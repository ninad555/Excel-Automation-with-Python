```python
path1 = "BPC_report-raw -2021.xlsx"
wb = load_workbook(path1)

ws = wb['YT']
print(ws.max_row, ws.max_column)
```

```python
end=ws.cell(ws.max_row,ws.max_column).column_letter + str(ws.cell(ws.max_row,ws.max_column).row)
start = "A2"
```

```python
formula_col_list = [] 
unique_col_list = []
for row in ws[start:end]:
    for cell in row:
        if cell.data_type == "f":
            formula_col_list.append(cell.column_letter)
for each_col in formula_col_list:
    if each_col not in unique_col_list:
        unique_col_list.append(each_col)
print(unique_col_list)
```

```python
for each_unique_col in unique_col_list:
    for row in ws['{}1:{}{}'.format(each_unique_col,each_unique_col,ws.max_row)]:
        for cell in row:
                ws["{}{}".format(cell.column_letter,cell.row)] = "=SUM(I{}:M{})".format(cell.row, cell.row)

wb.save("BPC_report-raw -2021.xlsx")
```
